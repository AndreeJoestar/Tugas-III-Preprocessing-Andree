```python

import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder

df = pd.read_csv("books_dataset.csv")

print("DATA AWAL")
display(df.head())

print("\nINFO DATA")
df.info()

print("\nCEK MISSING VALUE")
print(df.isnull().sum())

df['price'] = df['price'].astype(float)

print("\nSETELAH CLEANING PRICE")
display(df.head())

df['title_length'] = df['title'].apply(len)

df['word_count'] = df['title'].apply(lambda x: len(x.split()))

label_encoder = LabelEncoder()
df['availability_encoded'] = label_encoder.fit_transform(df['availability'])

scaler = MinMaxScaler()
df['price_scaled'] = scaler.fit_transform(df[['price']])

df['contains_history'] = df['title'].str.contains("history", case=False).astype(int)
df['contains_love'] = df['title'].str.contains("love", case=False).astype(int)
df['contains_science'] = df['title'].str.contains("science", case=False).astype(int)

print("\nHASIL FEATURE ENGINEERING")
display(df.head())

df.to_csv("books_dataset_processed.csv", index=False)
print("\nFILE DISIMPAN:books_dataset_processed.csv")
```
